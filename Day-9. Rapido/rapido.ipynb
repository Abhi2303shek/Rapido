### Q48. Which vehicle type gives the best revenue-to-cancellation ratio?

```python
revenue_by_vehicle = rapido.groupby('vehicle')['value'].sum()
cancel_count = rapido[rapido['booking_status']!='Success'].groupby('vehicle').size()

(revenue_by_vehicle / cancel_count).sort_values(ascending=False)
```

### Q49. At what ride distance does booking value increase sharply?


```python
successful.groupby(pd.cut(successful['dist'], bins=[0,10,20,30,40,50], labels=['V.Short','Short','Medium','Long','Longest']))['value'].mean()
```

### Q50. Which combination of pickup location + vehicle type is most profitable?

### Q51. Can peak-hour surge be identified from booking value patterns?


```python
successful['hour'] = pd.to_datetime(successful['time']).dt.hour
```

```python
successful.groupby((pd.cut(successful['hour'], bins=[0,4,8,12,16,20,24], labels=['Late','Early','Peak1','Afternoon','Peak2','LateNight'])))['value'].mean().sort_values(ascending=False)
```
